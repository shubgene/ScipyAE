```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
iris=load_iris()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
label = pd.DataFrame(list(map(lambda x : iris.target_names[x], iris.target)), columns=['Species'])
iris = pd.concat([data, label], axis=1) 
print(iris.head())
```

```python
plt.subplot(2,2,1)
sns.distplot(iris['sepal length (cm)'],color='y')
plt.subplot(2,2,2)
sns.distplot(iris['sepal width (cm)'],color='b')
plt.subplot(2,2,3)
sns.distplot(iris['petal length (cm)'],color='r')
plt.subplot(2,2,4)
sns.distplot(iris['petal width (cm)'],color='g')
```

```python
sns.boxplot(data=iris,palette='rainbow',orient='h')
```

```python
sns.countplot(x='Species',data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.lmplot(x='sepal length (cm)',y='petal length (cm)',hue="Species",data=iris)
```

```python
sns.barplot(x="Species",y='sepal length (cm)',data=iris)
```

```python
corr=iris.corr()
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```
